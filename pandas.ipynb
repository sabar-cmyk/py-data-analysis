#### dataframe 二维数据 行索引+ 列名  各列可不同数据类型  整张excel工作表
#### pd.DataFrame({'col',[1,2,3]})

```python
import pandas as pd
#自定义索引
s=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
print(s)
#定义标签
s=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'],name='my_series')
print(s)
#通过字典来创建
s=pd.Series({'a':1,'b':2,'c':3,'d':4,'e':5})
print(s)    
#通过索引去访问元素
s1=pd.Series(s,index=['a','c'])
print(s1)
```
