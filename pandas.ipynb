```python
np.random.seed(44)
score=pd.Series(np.random.randint(50,101,size=10),index=['学生'+str(i) for i in range(1,11)],name='score')
print(score)
```

```python
print(score.mean())  #平均值
print(score.max())  #最大值
print(score.min())  #最小值 
print((score>=score.mean()).sum())  #统计大于平均值的元素个数
s1=score[score>=score.mean()]  #访问大于平均值的元素
print(s1)  #访问大于平均值的元素
```

```python
#找出学生成绩变化最大的学生
score_diff=score.diff()  #计算成绩的差分
  #按值排序
print(score_diff)  #查看成绩变化
print(score_diff.sort_values(ascending=False)[0:2])  #查看成绩变化最大的前两名
```

```python
price=pd.Series([102.3,103.5,105.1,104.8,106.2,107.0,106.5,108.1,109.3,110.2],index=pd.date_range('2024-01-01', periods=10),name='price')
print(price)
```

```python
#计算价格的日收益率 当日收盘价-前日收盘价 -1
returns=price.pct_change()  #计算日收益率
returns.fillna(0, inplace=True)  #填充第一个元素为0
print(returns)  #查看日收益率
print('最大日收益率: {:.2f}%'.format(returns.max()*100))  #查看最大日收益率    
print('最小日收益率: {:.2f}%'.format(returns.min()*100))  #查看最小日收益率
print('日收益率标准差: {:.2f}%'.format(returns.std()*100))  #查看日收益率的标准差
print('最大日收益率日期: {}'.format(returns.idxmax()))  #查看最大日收益率的日期
print('最小日收益率日期: {}'.format(returns.idxmin()))  #查看最小日收益率的日期    
```

```python
sales=pd.Series([120,135,145,160,155,170,180,175,190,200,210,220]
                ,index=pd.date_range('2024-01-01', periods=12,freq='MS'),name='sales')#freq ='me'是月尾 ‘W’按周
print(sales)# 每个月份的销售额
```

```python
#季度的平均销量
print(sales.resample('QS').mean()) #重新采样' YS'年初 ‘QS’季度初
print(sales.resample('QS').sum()) #重新采样
```

```python
#最大销量
print(sales.idxmax(),sales.max())
#月环比的增长率
a=sales.pct_change()
b=a>0
b[b.rolling(3).sum()==3].keys().tolist()#滑动窗口
```

## DataFrame



```python
#dataframe_create
import pandas as pd
import numpy as np
# 通过series创建dataframe
s1=pd.Series([1,2,3],index=['a','n','c'],name='s1')
s2=pd.Series([4,5,6],index=['a','b','c'],name='s2')
df=pd.DataFrame({'First':s1,'Second':s2})  
print(type(df))  #查看数据类型
print(df)   
```

```python
#通过字典来创建
df=pd.DataFrame({"id":[1,2,3,4,5],"name":['Alice','Bob','Charlie','David','Eve'],
                 "age":[25,30,35,40,45],"score":[80.4,85.2,90.5,88.0,92.3]},
                index=['a','b','c','d','e'],columns=['id','name','score','age'])  #指定索引
print(df)
```

## DataFrame 的属性

```python
print(df.index)  #查看索引
print(df.columns)  #查看列标签
print(df.values)  #查看数据值
```

```python
print(df.ndim)  #查看维度
print(df.shape)  #查看形状  
print(df.dtypes)  #查看每列的数据类型
print(df.size)  #查看元素个数
```

```python
#行列转置
print(df.T)
print(df.T.index)
```
