## 数据的清洗和预处理

```python
#缺失值的处理
# nan :缺失值的标志，表示数据缺失或不可用
# dropna() :删除包含缺失值的行或列  
# fillna() :用指定的值替换缺失值
# isna() :判断是否为缺失值，返回布尔值  
# df.dropna()  #删除包含缺失值的行
# df.dropna(axis=1)  #删除包含缺失值的列
# df.fillna(0)  #用0替换缺失值
```

```python
print(df.isna().sum())  # 统计每列缺失值的数量
```

```python
#统计每行缺失值的数量
print(df.isna().sum(axis=1).sort_values(ascending=False))  # axis=1表示按行统计
```

```python
df=df.dropna(how='all')  #删除全为缺失值的行 也可以是any表示删除至少有一个缺失值的行
df=df.dropna(thresh=3)  #如果有至少3个不是缺失值就保留
df=df.dropna(subset=['model'])  #删除price列中缺失值的行

print(df.isna().sum()) 
print(df.shape)
```

#### 类别型特征建议用 'unknown' 或众数填充。

数值型特征建议用均值/中位数填充，或用更复杂的插补方法。

不建议直接带nan训练，除非用能原生支持缺失值的模型（如XGBoost、LightGBM）。

```python
#填充缺失值
df.fillna({'bodyType ':'unknown','fuelType ':'unknown','gearbox ':'unknown'}
          , inplace=True)  #用'unknown'填充bodyType和fuelType列的缺失值
print(df.isna().sum())
```

```python
df.ffill(inplace=True)  #用前一个非缺失值填充缺失值
df.bfill(inplace=True)  #用后一个非缺失值填充缺失值
print(df.isna().sum())
```
